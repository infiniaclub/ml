```python
from sklearn.cluster import DBSCAN
import numpy as np
import pandas as pd
import seaborn as sns

x = np.array([3, 4, 5, 7, 8, 5, 8, 9, 3, 3, 2, 5])
y = np.array([7, 6, 5, 5, 4, 1, 3, 3, 7, 5, 4, 9])

data=np.column_stack((x,y))
print(data)

model=DBSCAN(eps=1.9,min_samples=4)
model.fit(data)

sns.scatterplot(x=x,y=y,hue=model.labels_,palette="coolwarm")
```
